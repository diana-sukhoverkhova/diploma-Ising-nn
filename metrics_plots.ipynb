```python
Jd = -0.3
Ls = [10, 20, 30, 60, 80]
accuracy = {L: [] for L in Ls}
mse = {L: [] for L in Ls}
logloss = {L: [] for L in Ls}

if Jd == -0.5:
    L = 120
    accuracy[L] = []
    mse[L] = []
    logloss[L] = []

for epoch in range(1, 31):
    for L in Ls:
        elem = metrics[epoch-1][L][idxs[Jd]] ### [0] --- Jd == 0.0
        accuracy[L].append(elem[1])
        mse[L].append(elem[2])
        logloss[L].append(elem[3])
    if Jd == -0.5:
        L = 120
        with open(f'metrics/{epoch}_epoch_120.npy', 'rb') as f:
            metrics_120 = np.load(f, allow_pickle=True).item()
        accuracy[L].append(metrics_120[L][0][1])
        mse[L].append(metrics_120[L][0][2])
        logloss[L].append(metrics_120[L][0][3])
```
